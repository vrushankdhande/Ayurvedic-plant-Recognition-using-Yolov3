#Creating classes, labelled data, train and test file.
1. creating putting the count of classes in the labelled data
2. storing the path in labelled data


```python
path_to_images = '/content/drive/MyDrive/yol/final1'                                                   #path of were images are present 
count = 0

with open(path_to_images + '/' + 'classes.names', 'w') as names,                                       #creating files and counting classes by opening classes.txt
     open(path_to_images + '/' + 'classes.txt', 'r') as txt:
    for lines in txt:
        names.write(lines) 
        count += 1
        
with open(path_to_images + '/' + 'labelled_data.data', 'w') as data:
    data.write('classes = ' + str(count) + '\n')                                                      #tag name and count of classes (/n for next line)

    data.write('train = ' + path_to_images + '/' + 'train.txt' + '\n')                                # path of the train.txt file

    data.write('valid = ' + path_to_images + '/' + 'test.txt' + '\n')                                 # path of the test.txt file

    data.write('names = ' + path_to_images + '/' + 'classes.names' + '\n')                            # path of the classes.names file                   
  
    data.write('backup = backup')                                                                     # path where to save weights , keep this in the main folder of the main
                                                                                                      # yolo folder
```

# Divding all the images in 85 and 15 ratio for traing and testing data
1. taking all path of images and storing in a list
2. dividing and open the files train.txt and test.txt
3. writing all the path with ratio and saving it.

```python
full_path_to_images = '/content/drive/MyDrive/yol/final1'                                     # path of were images are present                        
os.chdir(full_path_to_images)
p = []

for current_dir, dirs, files in os.walk('.'):                                                 # Going through all files   
    for f in files:
        if f.endswith('.jpg'):                                                                # Checking if files with ends with '.jpg'(or what iamges format u have 'jpeg,etc')   
            path_to_save_into_txt_files = full_path_to_images + '/' + f
            p.append(path_to_save_into_txt_files + '\n')

p_test = p[:int(len(p) * 0.15)]                                                               #dividing in 15
p = p[int(len(p) * 0.15):]                                                                    #diving in 85

with open('train.txt', 'w') as train_txt:                                                     # Creating file train.txt and writing 85% of paths of images 
    for e in p:
        train_txt.write(e)

with open('test.txt', 'w') as test_txt:                                                       # Creating file test.txt and writing 15% of paths of images
    for e in p_test:
        test_txt.write(e)
```
